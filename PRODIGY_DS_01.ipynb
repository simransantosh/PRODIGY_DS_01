```python
import matplotlib.pyplot as plt
import numpy as np

# Filter data for the chosen country
chosen_country = 'United States'
country_data = df_final[df_final['Country Name'] == chosen_country]

# Extract years and population data
years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

# Create figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart - Population over the years
ax1.bar(years, population, color='skyblue', label='Population')
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color='skyblue')
ax1.set_title(f'Population Trend and Distribution of {chosen_country}')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Histogram - Distribution of Population
ax2 = ax1.twinx()
ax2.hist(population, bins=20, color='lightgreen', edgecolor='black', alpha=0.7, label='Population Distribution')
ax2.set_ylabel('Frequency', color='lightgreen')
ax2.tick_params(axis='y', labelcolor='lightgreen')

# Add legends
fig.legend(loc='upper left')

plt.show()
```


In this visualization, the bar chart represents the population trend over the years, while the histogram displays the distribution of population data. By combining both plots, we can gain insights into both the overall trend and the distribution of population data for the chosen country, United States, in a single visualization.





**Bar Chart (Population over the Years):**

The bar chart represents the population trend over the years for the chosen country, which in this case is the United States.
The x-axis represents the years, while the y-axis represents the population.
Each bar represents the population count for a specific year.
The bars are colored sky blue for better visibility.

**Histogram (Distribution of Population):**

The histogram displays the distribution of population data for the chosen country.
It represents how frequently different population ranges occur within the dataset.
The x-axis represents different population ranges, divided into bins. In this case, 20 bins are used.
The y-axis represents the frequency or count of data points falling within each bin.
The histogram bars are colored light green with black edges.

**Dual-Axis Plot:**

The bar chart and histogram are combined into a single visualization using a dual-axis plot.
The bar chart and histogram share the same x-axis (years) but have different y-axes, allowing for easy comparison between the population trend and its distribution.
The left y-axis corresponds to the population count in the bar chart, while the right y-axis corresponds to the frequency of population ranges in the histogram.
The legends on the upper left corner indicate which line corresponds to which type of data (population trend or population distribution).

**Overall Interpretation:**

By examining the combined visualization, we can observe how the population of the United States has changed over the years.
The bar chart provides insight into overall population trends, such as periods of growth or decline.
The histogram complements this by showing the distribution of population data, highlighting the concentration of population within certain ranges.
Together, these visualizations offer a comprehensive understanding of both the population trend and its distribution for the chosen country.
